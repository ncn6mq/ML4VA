```python
def fixEthnicity(adict):
    ethnicities = adict["ethnicityInfo"]
    for ethnicity in ethnicities:
        if ethnicity['label'] != 'All students':
            adict[ethnicity['label']] = ethnicity['percentage']
```

```python
import numpy as np

def mergeDict(dict1, dict2add):
    for key in dict1.keys():
        if key in dict2add:
            dict1[key].append(dict2add[key])
        else:
            dict1[key].append(np.nan)

```

```python
import json
from bs4 import BeautifulSoup
import pandas as pd
import requests

example = "https://www.greatschools.org/virginia/schools/?gradeLevels%5B%5D=h&gradeLevels%5B%5D=m&gradeLevels%5B%5D=e&page=1&st%5B%5D=public_charter&st%5B%5D=public&st%5B%5D=charter&view=table"

page = requests.get(example)

soup = BeautifulSoup(contents, 'html.parser')
raw_text = soup.find("script").get_text().split(";")
js = "gon.search="

df = dict()

for i in range(len(raw_text)):
    piece = raw_text[i]
    if piece[:11] == js:
        obj = piece.replace(js, "")
        obj = json.loads(obj)
        data = obj["schools"]
        for datum in data:
            fixEthnicity(datum)
            df = mergeDict(df, datum)
```

```python
import requests

base = "https://www.greatschools.org/virginia/schools/?gradeLevels%5B%5D=h&gradeLevels%5B%5D=m&gradeLevels%5B%5D=e&page="
end = "&st%5B%5D=public_charter&st%5B%5D=public&st%5B%5D=charter&view=table"

schools = list()

js = "gon.search="

for i in range(1, 85):
    page = requests.get(base + str(i) + end)
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    raw_text = soup.find("script").get_text().split(";")
    for i in range(len(raw_text)):
        piece = raw_text[i]
        if piece[:11] == js:
            obj = piece.replace(js, "")
            obj = json.loads(obj)
            data = obj["schools"]
            schools += data
```

```python
keys = list()
```

```python
for school in schools:
    fixEthnicity(school)
```

```python
for school in schools:
    for key in school.keys():
        keys.append(key)
```

```python
pd.Series(keys).value_counts()
```

```python
df = dict()
for key in keys:
    df[key] = []
```

```python
mergeDict(df, schools[1])
```

```python
for school in schools:
    mergeDict(df, school)
```

```python
df = pd.DataFrame(df)
```
