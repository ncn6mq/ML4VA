```python
salaries = pd.read_csv("teachers.csv")
dems = pd.read_csv("demographic.csv")
scores = pd.read_csv("school-by-subject-2019.csv")
```

```python
salaries
```

```python
dems
```

```python
scores
```

```python
scores["Div Name"] = scores["Div Name"].apply(lambda x: x + 'Public Schools')
```

```python
salaries = salaries.set_index(salaries["Div Name"])
salaries.drop("Div Name", axis='columns', inplace=True)
```

```python
dems = dems.set_index(dems["Div Name"])
dems.drop(["Div Name", "GeoId", "LEAID", "Year", "Iteration"], axis='columns', inplace=True)
```

```python
scores = scores.join(salaries, how='left', on='Div Name')
```

```python
scores = scores.join(dems, how='left', on='Div Name')
```
