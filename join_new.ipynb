```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 99)
```

```python
df = pd.read_csv('ScoresSalsDems.csv').drop("Unnamed: 0", axis=1)
```

```python
df
```

```python
names1 = list(df.iloc[:, :13].columns)
```

```python
names2 = list(df.iloc[:, 13:].columns)
```

```python
for i in range(len(names2)):
    names2[i] = names2[i].replace(" ", "") + "Dist"
```

```python
df.columns = names1 + names2
```

```python
by_school = pd.read_csv('by_school.csv').drop("Unnamed: 0", axis=1)
```

```python
by_school.columns = ['Div Name', 'DistrictCity', 'Latitude', 'Longitude', 'Sch Name', 'Sch Type',
       'Rating', 'PercentLowIncome', 'Enrollment', 'ParentRating',
       'NumReviews', 'PercentAsianSch', 'PercentBlackSch', 'PercentHispanicSch', 'PercentWhiteSch',
       'PercentMultiethSch', 'PercentNativeSch',
       'StudentsPerTeacher', 'PercentIslanderSch']
```

```python
by_school = by_school.drop('Sch Type', axis=1)
```

```python
by_school["PercentAsianSch"] = by_school["PercentAsianSch"].fillna(0)
by_school["PercentBlackSch"] = by_school["PercentBlackSch"].fillna(0)
by_school["PercentHispanicSch"] = by_school["PercentHispanicSch"].fillna(0)
by_school["PercentWhiteSch"] = by_school["PercentWhiteSch"].fillna(0)
by_school["PercentMultiethSch"] = by_school["PercentMultiethSch"].fillna(0)
by_school["PercentNativeSch"] = by_school["PercentNativeSch"].fillna(0)
by_school["PercentIslanderSch"] = by_school["PercentIslanderSch"].fillna(0)
```

```python
by_school['PercentLowIncome'] = by_school['PercentLowIncome'].apply(lambda x: np.nan if x == '?' else x.replace("%", "")).astype(float)
```

```python
wrong_zero = ["Enrollment", "ParentRating", "NumReviews"]

for col in wrong_zero:
    by_school.loc[by_school[col] == 0, col] = np.nan
```

```python
by_school["MissingRatio"] = by_school["StudentsPerTeacher"].isnull()
by_school["MissingEnrollment"] = by_school["Enrollment"].isnull()
```

```python
from sklearn.impute import SimpleImputer

to_impute = ["PercentLowIncome", "Enrollment", "Rating", "ParentRating", "NumReviews", "StudentsPerTeacher"]
```

```python
imputer = SimpleImputer(strategy='mean')

fitted = imputer.fit(by_school.loc[:, to_impute])
by_school.loc[:, to_impute] = imputer.transform(by_school.loc[:, to_impute])
```

```python
by_school.isna().sum()
```

```python
by_school = by_school.loc[~by_school["Div Name"].isna(), :]
```

```python
by_school["MissingRatio"] = by_school["MissingRatio"].astype(float)
by_school["MissingEnrollment"] = by_school["MissingEnrollment"].astype(float)
```

```python
test = by_school.copy()

test["Sch Name"] = test["Sch Name"].str.replace(" School", "")
test["Sch Name"] = test["Sch Name"].str.replace(".", "")
test["Sch Name"] = test["Sch Name"].str.replace(" ", "")
#df["Sch Name"] = df["Sch Name"].str.replace(" School", "")
df["Sch Name"] = df["Sch Name"].str.replace(" ", "")
df["Sch Name"] = df["Sch Name"].str.replace(".", "")
```

```python
test = test.set_index(["Div Name", "Sch Name"])

findna = df.join(test, how='left', on=["Div Name", "Sch Name"], rsuffix='test')
```
